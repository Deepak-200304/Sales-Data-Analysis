```python
import pandas as pd

df = pd.read_csv('sales_data.csv')

df.head()
```

```python
df.fillna(0, inplace=True)

df.drop_duplicates(inplace=True)

df.isnull().sum()
```

```python
if 'Total_Sales' not in df.columns:
    df['Total_Sales'] = df['Quantity'] * df['Price']

df.head()
```

```python
total_revenue = df['Total_Sales'].sum()

best_selling_product = df.groupby('Product')['Total_Sales'].sum().idxmax()

average_sales = df['Total_Sales'].mean()

total_revenue, best_selling_product, average_sales
```

```python
print("SALES DATA ANALYSIS REPORT")
print("-" * 40)

print(f"Total Revenue          : ₹{total_revenue:,.2f}")
print(f"Best-Selling Product   : {best_selling_product}")
print(f"Average Sales Value    : ₹{average_sales:,.2f}")
```

```python
print("\n KEY INSIGHTS")
print("-" * 40)
print("• The dataset was cleaned by handling missing values and duplicates.")
print(f"• '{best_selling_product}' generated the highest revenue.")
print("• Average sales value helps understand customer purchase behavior.")
```
